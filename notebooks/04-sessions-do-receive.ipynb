# DO Phase 2: Resume, receive processed data, and verify restore

```python
import beaver

session = beaver.active_session()
assert session is not None, 'Session should be available'

# Load prior published state
session.load(overwrite=True)
assert 'phase' in globals() and phase == 'published', 'Expected published phase'

processed_data = session.wait_for_remote_var('processed_data', auto_accept=True)
if hasattr(processed_data, 'load'):
    processed_data = processed_data.load(auto_accept=True)
elif hasattr(processed_data, 'value'):
    processed_data = processed_data.value
phase = 'received'
state_path = session.save()
print('Received processed data:', processed_data)
print(f'State saved to: {state_path}')
# Simulate restart and ensure globals are restored
processed_data = None
phase = None
session.load(path=state_path, overwrite=True)
assert processed_data == 14
assert phase == 'received'
print('State restored and verified after restart')
```
